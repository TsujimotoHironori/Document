```python
# ライブラリーインポート、グラフ描画準備
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import re
#plt.rcParams['font.family'] = 'IPAexGothic'

# matplotlib inline
plt.style.use('ggplot')
```

```python
# PandasでCSVファイルを読み込む
#parse_dates=[0]で0列目をdatetime型へ変換

filepass = r'C:\Users\4440\Desktop\TG117264\AI特許分析モデル\7.tms_to_python\tms_to_python_内視鏡.csv'

read_csv = pd.read_csv(filepass,parse_dates=[0],encoding='Shift_JIS')
#print(df1)
read_csv = read_csv[read_csv['品詞'].isin(['名詞'])]
read_csv = read_csv[~read_csv['品詞詳細'].isin(['数詞','副詞可能','固有名詞人名'])]
```

```python
#指定列を抽出
df = read_csv[['行ID','置換語']]

#リストを作成し、列追加
l1 = []
l2 = []*2

#l1はidが同じ場合に置換語を代入
#idが変更する際にl2へ代入
#正規表現で数値は削除
id = 1
for index, row in df.iterrows():
    if row['行ID'] == id:
        word = row['置換語']
        word_re = re.sub('[\d]','',word)
        l1.append(word_re)
    else:
        l2.append(l1)
        #print(l1)
        l1 = []
        word = row['置換語']
        word_re = re.sub('[\d]','',word)
        l1.append(word_re)
        id += 1 
l2.append(l1)        
print(len(l2))
```

```python
#ファイルへ書き込み
file = open(r'C:\Users\4440\Desktop\TG117264\AI特許分析モデル\7.tms_to_python\tms_to_python_内視鏡.txt', 'a')  #書き込みモードでオープン
for i in range(len(l2)):
    string = ""
    for n in range(len(l2[i])):
        string += l2[i][n] + " "
    file.write(string)
    file.write('\n')
file.close()
```
