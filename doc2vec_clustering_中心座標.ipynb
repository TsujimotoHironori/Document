```python

#coding: UTF-8

#ベクトル化手法
import numpy as np
import pandas as pd
import matplotlib.dates as mdates
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
import sys
from collections import defaultdict
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA        # 今回の主役。scikit-learnのPCAクラス
from sklearn import datasets                 # デモ用のデータセットを読み込むモジュール
import seaborn as sns                        # 可視化用のライブラリ
import pandas as pd                          # 便利なDataFrameを使うためのライブラリ
from gensim.models.doc2vec import Doc2Vec

f = open(r'C:\Users\4440\Desktop\k-means\tms_to_python_カメラ.txt','r')
#txtファイルを配列に格納
trainings = []
for doc in f:
    trainings.append(doc)


#BoW実行
m = Doc2Vec.load('doc2vec_カメラ.model')
vectors_list=[m.docvecs[n] for n in range(len(m.docvecs))]

#ベクトルをリストに格納
print(vectors_list)

df = pd.DataFrame(vectors_list)
print(df)
```

```python
#ドキュメント番号のリスト
doc_nums=range(1,len(m.docvecs))

#クラスタリング設定
#クラスター数を変えたい場合はn_clustersを変えてください
n_clusters = 8
kmeans_model = KMeans(n_clusters=n_clusters,init="k-means++", verbose=1, random_state=1, n_jobs=-1)

#クラスタリング実行
kmeans_model.fit(vectors_list)

#クラスタリングデータにラベル付け
labels=kmeans_model.labels_

#ラベルとドキュメント番号の辞書づくり
cluster_to_docs = defaultdict(list)
for cluster_id, doc_num in zip(labels, doc_nums):
    cluster_to_docs[cluster_id].append(doc_num)

#クラスター出力
for docs in cluster_to_docs.values():
    print(docs)
```
